```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

## Load Data [Hotel booking demand](https://www.kaggle.com/datasets/jessemostipak/hotel-booking-demand) Dataset from JESSE MOSTIPAK

```python
hotel_booking = pd.read_csv('/kaggle/input/hotel-booking-demand/hotel_bookings.csv')
```

```python
hotel_booking.info()
```

```python
hotel_booking.isnull().sum()
```

## Manipulate data

```python
hotel_booking['children'] = hotel_booking['children'].fillna(0)
```

```python
hotel_booking['country'].mode()
```

```python
hotel_booking['country'] = hotel_booking['country'].fillna('PRT')
```

```python
hotel_booking['agent'] = hotel_booking['agent'].fillna(hotel_booking['agent'].mean())
```

```python
pd.unique(hotel_booking['company'])
```

```python
hotel_booking['company'] = hotel_booking['company'].fillna('regular customer')
```

```python
hotel_booking.isnull().sum()
```

```python
hotel_booking['company'] = hotel_booking['company'].astype(str)
```

```python
hotel_booking.info()
```

```python
hotel_booking.to_csv('hotel_booking.csv')
```

```python
hotel_booking.head()
```

## Create insigt chart

```python
import seaborn as sns
import matplotlib as plt
```

```python
colors = ['red','pink']
data = hotel_booking['hotel'].value_counts()
ax = data.plot(kind="pie", 
               autopct=lambda p: '{:.2f}%\n({:.0f})'.format(p,(p/100)*data.sum()), 
               shadow=True, 
               explode=[0.05, 0.05], 
               colors=colors, 
               legend=True, 
               title='Hotel Type', 
               ylabel='', 
               labeldistance=None)
ax.legend(bbox_to_anchor=(1, 1.02), loc='upper left')
```

```python
company_type = hotel_booking['company'].value_counts()
```

```python
company_type = pd.DataFrame({'company': company_type.index, 'count': company_type.values})
```

```python
type(company_type)
```

```python
company_type = company_type.sort_values(by='count', ascending=False)
```

```python
company_type = company_type.reset_index(drop=True)
```
